```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)



# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
#school_data.head()
#student_data.head()
```

```python
#Calculating the total number of schools without repetition
total_schools = len(school_data_complete["school_name"].unique())
```

```python
total_students = len(school_data_complete["student_name"] )# there were repeatable students names too
```

```python
#Before I calculate the total budget
#I have to calculate the average budget of the repeatable names of the schools.
df_budg = school_data_complete.groupby("school_name").mean()
#Calculate the total budget
total_budget = df_budg["budget"].sum() 
```

```python
# I wasn't sure whether I should calculate the average math score with the respect to districts or total schools.
# The calculation below, is the average of total of schools.
avg_ms = school_data_complete["math_score"].mean()
```

```python
avg_rs = school_data_complete["reading_score"].mean()
```

```python
overall_passing_rate = (avg_ms + avg_rs)/2
overall_passing_rate
```

```python
# the calculation is based on the assumption that students are "unique" individuals, regardless of their names similarity.
total_students_all =  len(school_data_complete["student_name"])
total_students_all
```

```python
# percentage = (total number of students with 70 or greater score/ total number of students)*100. 
school_data_complete["math_score"]>=70
```

```python
# When I run the above I got all the values for all math scores above or equal 70, now I can get the total number
students_above_70 = sum(school_data_complete["math_score"]>=70)
```

```python
#calculating the percentage
pct_math = (students_above_70/total_students_all)*100
pct_math
```

```python
# Calculating the percentage of students with a passing reading score (70 or greater)
school_data_complete["reading_score"]>=70
students_above_70_r = sum(school_data_complete["reading_score"]>=70)
pct_reading = (students_above_70_r/total_students_all)*100
pct_reading
```

```python
#Create a dataframe to hold the above results
# created dictionary first to make it easier for data frame
d = {"Total_Schools": total_schools ,
     "Total_students": total_students ,
     "Total_budget" : total_budget ,
    "average_math_score " : avg_ms , 
     "average_reading_score" : avg_rs ,
     "overall_passing_rate": overall_passing_rate,
    "passing_math_score" : pct_math,
    "passing_reading_score" : pct_reading}

df_results = pd.DataFrame(data = d , index=[0])# error message occured until the index was inserted
df_results
```

```python
# by using groupby I was able to answer most of the questions, I still need info like school type 
#sc_type = school_data_complete.set_index("school_name").groupby("school_name")["type"]

type_schools = school_data.set_index("school_name")["type"]
students_schools = school_data.set_index("school_name")["size"]
budget_schools = school_data.set_index("school_name")["budget"]
per_student = budget_schools/students_schools
math_avg = school_data_complete.groupby("school_name")["math_score"].mean()
reading_avg = school_data_complete.groupby("school_name")["reading_score"].mean()

studentsprc_above_70_math = (school_data_complete[school_data_complete["math_score"]>=70].groupby("school_name")["Student ID"].count()/students_schools)*100

studentsprc_above_70_reading = (school_data_complete[school_data_complete["reading_score"]>=70].groupby("school_name")["Student ID"].count()/students_schools)*100

overall_passing = (studentsprc_above_70_math +studentsprc_above_70_reading)/2

```

```python
#Create a dataframe to hold the above results

d = {"School type": type_schools,
     "Total_students": students_schools ,
     "Total_budget" : budget_schools ,
     "Per_student_budget": per_student,
    "average_math_score " : math_avg , 
     "average_reading_score" : reading_avg ,
     "overall_passing_rate": overall_passing,
    "passing_math_score" : studentsprc_above_70_math,
    "passing_reading_score" : studentsprc_above_70_reading}

df_results_school = pd.DataFrame(data = d )
df_results_school

```
